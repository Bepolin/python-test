```python
from datetime import datetime, date, timedelta
import pandas as pd
import plotly.express as px
```

```python
date = datetime(2021, 12, 31)
stock = 123133
sale1 = 120
sale2 = 100
print(date)
```

```python
data = []
for i in range(1, 10000):
    new_data = date + timedelta(days=i)
    new_sale1 = stock-i*sale1
    new_sale2 = stock-i*sale2
    if new_sale1 < 0 and new_sale2 < 0:
        break
    data.append([new_data, new_sale1, new_sale2])
print(len(data))
print(data[-1])
```

```python
df = pd.DataFrame(data, columns=['date', 'sale1', 'sale2'])
print(df.head())
fig = px.line(df, x="date", y="sale1")
fig.show()
```
